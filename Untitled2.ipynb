```python
import matplotlib.pyplot as plt

# Define both lists
month = ['Jan', 'Feb', 'Mar', 'Apr']
sales = [200, 300, 400, 500]

# Create bar chart
plt.bar(month, sales)
plt.xlabel('<-----Month----->')
plt.ylabel('<------Sales----->')
plt.title('Sales per Month')
plt.show()

```
